```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 3
x = np.linspace(-10,10,21)

a = st.poisson(mu)

fig,ax = plt.subplots(3,1)
ax[0].plot(x,a.pmf(x))
ax[1].plot(x,a.cdf(x))
ax[2].hist(a.rvs(1000), bins=x)
```

```python
b = st.norm(mu)

fig,ax = plt.subplots(3,1)
ax[0].plot(x,b.pdf(x))
ax[1].plot(x,b.cdf(x))
ax[2].hist(b.rvs(1000), bins=x)
```

```python
c1 = st.norm(3)
c2 = st.norm(3)

sample_1a = c1.rvs(200) + 1
sample_1b = c1.rvs(200) + 1
sample_2 = c2.rvs(200)

fig, ax = plt.subplots(1, 1)
ax.hist(sample_1a)
ax.hist(sample_2)
```

```python
print(st.ttest_ind(sample_1a,sample_2))
```

```python
print(st.ttest_ind(sample_1a,sample_1b))
```
